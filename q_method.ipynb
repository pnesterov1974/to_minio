```python
import os
from polished import pmain
import pandas               as pd
import nb_utils             as nbu

p = pmain.create_polished( 'postgresql' )
pc  = p.prrf_current
r   = p.rapaport
q   = p.quotations
rg  = p.regrading
mst = p.masters
rg.make_work_tables()
```

```python
df_index, df_data, df_cols = rg.prrf_base_df
df_prrf_base = pd.DataFrame( df_data, columns=df_cols, index=df_index )
df_prrf_base[ 'prrfcut_name' ] = df_prrf_base.apply( lambda row: mst.get_prrfcut_name(row[ 'prrfcut_id' ]), axis=1 )
df_prrf_base[ 'russzgrp_name' ] = df_prrf_base.apply( lambda row: mst.get_russzgrp_name(row[ 'russzgrp_id' ]), axis=1 )
df_prrf_base[ 'ruscolor_name' ] = df_prrf_base.apply( lambda row: mst.get_ruscolor_name(row[ 'ruscolor_id' ]), axis=1 )
df_prrf_base[ 'rusclarity_name' ] = df_prrf_base.apply( lambda row: mst.get_rusclarity_name(row[ 'rusclarity_id' ]), axis=1 )
df_prrf_base[ 'qrusclaritygrp_name' ] = df_prrf_base.apply( lambda row: mst.get_qrusclaritygrp_name(row[ 'qrusclaritygrp_id' ]), axis=1 )
df_prrf_base[ 'cprice' ] = df_prrf_base.apply( lambda row: row[ 'pprice' ] * row[ 'factor' ], axis=1 )

#df_prrf_base = df_prrf_base[ [ 'prrfcut_name', 'prrfcut_id', 'russzgrp_name', 'russzgrp_id', 'ruscolor_name', 'ruscolor_id',
#                                'rusclarity_name','rusclarity_id', 'qrusclaritygrp_name', 'qrusclaritygrp_id', 'pprice', 'factor', 'cprice'
#                             ]
#                           ]
df_prrf_base = df_prrf_base[ [ 'prrfcut_name', 'russzgrp_name', 'ruscolor_name', 'rusclarity_name', 'qrusclaritygrp_name',
                               'pprice', 'factor', 'cprice'
                             ]
                           ]
#df_prrf_base
```

```python
nbu.pandas_to_xlsx(df_prrf_base, os.path.join('d:\\','py_dev','work','out_files','prrf_base.xlsx'), 'prrf_base')
```

```python
df_index, df_data, df_cols = rg.gia_table_df
df_gia_table = pd.DataFrame( df_data, columns=df_cols, index=df_index )

df_gia_table[ 'rapcut_name' ] = df_gia_table.apply( lambda row: mst.get_rapcut_name( row[ 'rapcut_id' ]), axis=1 )
df_gia_table[ 'giaszgrp_name' ] = df_gia_table.apply( lambda row: mst.get_giaszgrp_name( row[ 'giaszgrp_id' ]), axis=1 )
df_gia_table[ 'giacolor_name' ] = df_gia_table.apply( lambda row: mst.get_giacolor_name( row[ 'giacolor_id' ]), axis=1 )
df_gia_table[ 'giaclarity_name' ] = df_gia_table.apply( lambda row: mst.get_giaclarity_name( row[ 'giaclarity_id' ]), axis=1 )
df_gia_table[ 'russzgrp_from_name' ] = df_gia_table.apply( lambda row: mst.get_russzgrp_name( row[ 'russzgrp_from_id' ]), axis=1 )
df_gia_table[ 'russzgrp_to_name' ] = df_gia_table.apply( lambda row: mst.get_russzgrp_name( row[ 'russzgrp_to_id' ]), axis=1 )
df_gia_table[ 'ruscolor_from_name' ] = df_gia_table.apply( lambda row: mst.get_ruscolor_name( row[ 'ruscolor_from_id' ]), axis=1 )
df_gia_table[ 'ruscolor_to_name' ] = df_gia_table.apply( lambda row: mst.get_ruscolor_name( row[ 'ruscolor_to_id' ]), axis=1 )
df_gia_table[ 'rusclarity_from_name' ] = df_gia_table.apply( lambda row: mst.get_rusclarity_name( row[ 'rusclarity_from_id' ]), axis=1 )
df_gia_table[ 'rusclarity_to_name' ] = df_gia_table.apply( lambda row: mst.get_rusclarity_name( row[ 'rusclarity_to_id' ]), axis=1 )

df_gia_table = df_gia_table[ [ 'rapcut_name','giaszgrp_name','giacolor_name', 'giaclarity_name', 'russzgrp_from_name',
                               'russzgrp_to_name', 'ruscolor_from_name', 'ruscolor_to_name', 'rusclarity_from_name',
                               'rusclarity_to_name', 'pprice', 'cprice','rap100price'
                             ]
                           ]
df_gia_table
```
